### Atributos

- Permite resolver sistemas con más de 6 millones de grados de libertad en un computador portátil para elasticidad plana, armaduras y pórticos.

- Es un simulador. Para clases conceptuales donde el objetivo principal es la física.

- Es modular. Para clases procedimentales donde importan los métodos numéricos o los algoritmos.
